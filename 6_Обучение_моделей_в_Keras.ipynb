```python
#Давай сделаем полностью рабочий пример модели для MNIST с использованием Keras и TensorFlow. Код будет включать: загрузку данных, предобработку, построение модели, компиляцию, обучение и оценку.

# Импорт необходимых библиотек
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# 1. Загрузка данных MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Предобработка данных
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0

# Преобразуем метки в one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 3. Создание модели
model = Sequential([
    layers.InputLayer(input_shape=(28*28,)),  # входной слой
    layers.Dense(500, activation='relu'),     # первый скрытый слой
    layers.Dense(500, activation='relu'),     # второй скрытый слой
    layers.Dense(10, activation='softmax')    # выходной слой для 10 классов
])

# 4. Компиляция модели
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Обучение модели
model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1
)

# 6. Оценка модели на тестовых данных
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")


#✅ Ключевые моменты:

#input_shape=(28*28,) — обязательно кортеж.

#Метки one-hot encoded через to_categorical.

#Последний слой softmax для многоклассовой классификации.

#Используем categorical_crossentropy для функции потерь.
```

```python
#Давай сделаем расширенный рабочий пример для MNIST, который включает:

#Сохранение модели на диск после обучения.

#Построение графиков loss и accuracy для обучения и валидации, видеть кривые loss и accuracy — это удобно для анализа результатов.

#Вот полностью рабочий код:

# Импорт библиотек
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# 1. Загрузка данных MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Предобработка данных
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 3. Создание модели
model = Sequential([
    layers.InputLayer(input_shape=(28*28,)),
    layers.Dense(500, activation='relu'),
    layers.Dense(500, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Компиляция модели
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Обучение модели с возвратом истории обучения
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1
)

# 6. Сохранение модели на диск
model.save('mnist_model.h5')
print("Модель сохранена как 'mnist_model.h5'")

# 7. Оценка модели на тестовых данных
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")

# 8. Построение графиков loss и accuracy
plt.figure(figsize=(12, 5))

# Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()


#✅ Что делает этот код:

#Сохраняет модель в файл mnist_model.h5, который потом можно загрузить через tf.keras.models.load_model('mnist_model.h5').

#Строит графики loss и accuracy для тренировочной и валидационной выборки, чтобы сразу видеть качество обучения.
```

```python
#Давай добавим EarlyStopping и ModelCheckpoint, чтобы:

#Обучение автоматически останавливалось, если валидационная потеря перестанет улучшаться.

#Сохранялась лучшая модель по валидационной точности (accuracy).

#Вот полностью рабочий код с этими улучшениями:

# Импорт библиотек
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

# 1. Загрузка данных MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Предобработка данных
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 3. Создание модели
model = Sequential([
    layers.InputLayer(input_shape=(28*28,)),
    layers.Dense(500, activation='relu'),
    layers.Dense(500, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Компиляция модели
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Настройка колбэков
early_stop = EarlyStopping(
    monitor='val_loss',      # следим за валидационной потерей
    patience=3,              # ждём 3 эпохи без улучшений
    restore_best_weights=True # восстанавливаем лучшие веса
)

checkpoint = ModelCheckpoint(
    'best_mnist_model.h5',  # имя файла для лучшей модели
    monitor='val_accuracy',  # следим за точностью на валидации
    save_best_only=True,     # сохраняем только лучшую модель
    mode='max',              # ищем максимальное значение
    verbose=1
)

# 6. Обучение модели с колбэками
history = model.fit(
    x_train, y_train,
    epochs=50,               # можно задать больше эпох, обучение остановится автоматически
    batch_size=128,
    validation_split=0.1,
    callbacks=[early_stop, checkpoint]
)

# 7. Загрузка лучшей модели
best_model = tf.keras.models.load_model('best_mnist_model.h5')
print("Лучшая модель загружена из 'best_mnist_model.h5'")

# 8. Оценка лучшей модели на тестовых данных
test_loss, test_acc = best_model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")

# 9. Построение графиков loss и accuracy
plt.figure(figsize=(12, 5))

# Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()


#✅ Что добавлено:

#EarlyStopping: останавливает обучение при отсутствии улучшений на 3 эпохи.

#restore_best_weights=True: после остановки восстанавливаются лучшие веса.

#ModelCheckpoint: сохраняет лучшую модель по валидационной точности в файл best_mnist_model.h5.

#Можно увеличить epochs до 50 или больше — обучение остановится автоматически при переобучении.
```

Задание 6.1
1 point possible (graded)
Повторите процедуру создания и обучения модели 50 раз. Обучайте модель в течение 10 эпох. В каждом случае получите предсказания обученной модели на тестовых данных. Сохраните все предсказания в массив — они нам понадобятся далее. Чему равно медианное значение точности (accuracy) на тестовом датасете?

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Загрузка и предобработка данных
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0

y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# Массив для хранения точностей всех моделей
accuracies = []
# Массив для хранения предсказаний всех моделей
all_predictions = []

# Повторяем 50 раз
for i in range(50):
    # Создание модели
    model = Sequential([
        layers.InputLayer(input_shape=(28*28,)),
        layers.Dense(500, activation='relu'),
        layers.Dense(500, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    # Компиляция
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # Обучение 10 эпох без вывода
    model.fit(x_train, y_train_cat, epochs=10, batch_size=128, verbose=0)

    # Получение предсказаний на тесте
    preds = model.predict(x_test, verbose=0)
    all_predictions.append(preds)

    # Вычисление точности на тесте
    test_acc = np.mean(np.argmax(preds, axis=1) == y_test)
    accuracies.append(test_acc)

# Переводим списки в массивы для удобства
accuracies = np.array(accuracies)
all_predictions = np.array(all_predictions)  # форма (50, 10000, 10)

# Вычисление медианной точности
median_accuracy = np.median(accuracies)
print(f"Медианная точность на тестовом датасете: {median_accuracy:.4f}")
```

Цикл 50 раз создаёт новые модели с одинаковой архитектурой.

Каждая модель обучается 10 эпох с verbose=0, чтобы не засорять консоль.

Сохраняются предсказания каждой модели (all_predictions) для последующих задач.

 добавить GPU-ускорение и небольшую оптимизацию по памяти, чтобы 50 моделей обучались быстрее на ноутбуке.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# 0. Проверка доступности GPU
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# 1. Загрузка и предобработка данных
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0

y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# 2. Массивы для хранения результатов
accuracies = []
all_predictions = []

# 3. Оптимизация памяти: используем стратегию GPU и удаляем модели после использования
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        # Разрешаем динамическое выделение памяти GPU
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# 4. Цикл обучения 50 моделей
for i in range(50):
    # Создание модели
    model = Sequential([
        layers.InputLayer(input_shape=(28*28,)),
        layers.Dense(500, activation='relu'),
        layers.Dense(500, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    # Компиляция
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # Обучение 10 эпох
    model.fit(x_train, y_train_cat, epochs=10, batch_size=128, verbose=0)

    # Предсказания на тестовых данных
    preds = model.predict(x_test, verbose=0)
    all_predictions.append(preds)

    # Точность на тесте
    test_acc = np.mean(np.argmax(preds, axis=1) == y_test)
    accuracies.append(test_acc)

    # Удаляем модель и освобождаем память
    tf.keras.backend.clear_session()
    del model
    del preds

# 5. Преобразование в массивы
accuracies = np.array(accuracies)
all_predictions = np.array(all_predictions)  # форма (50, 10000, 10)

# 6. Вычисление медианной точности
median_accuracy = np.median(accuracies)
print(f"Медианная точность на тестовом датасете: {median_accuracy:.4f}")
```

✅ Что здесь улучшено:

GPU-ускорение: проверка доступности и динамическое выделение памяти (set_memory_growth) предотвращает переполнение памяти.

Оптимизация памяти:

После каждой модели вызываем tf.keras.backend.clear_session() и удаляем объекты model и preds.

Это позволяет обучать 50 моделей подряд без переполнения памяти.

Verbose=0 — обучение тихое, не засоряет консоль.

💡 Советы:

Создаём модель с той же архитектурой, но learning_rate=0.01 (1e-2).

Вот рабочий пример кода:

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# 1. Загрузка и предобработка данных
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0

y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# 2. Разделяем тестовый датасет на две части
x_test1, y_test1 = x_test[:5000], y_test[:5000]
x_test2, y_test2 = x_test[5000:], y_test[5000:]

# 3. Создание модели
model = Sequential([
    layers.InputLayer(input_shape=(28*28,)),
    layers.Dense(500, activation='relu'),
    layers.Dense(500, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Компиляция с learning_rate=0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-2)
model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Цикл обучения по 1 эпохе + оценка на двух тест-наборах
acc1_list = []
acc2_list = []

for epoch in range(100):
    # Обучение 1 эпохи
    model.fit(x_train, y_train_cat, epochs=1, batch_size=128, verbose=0)

    # Оценка на первой половине теста
    loss1, acc1 = model.evaluate(x_test1, y_test_cat[:5000], verbose=0)
    acc1_list.append(acc1)

    # Оценка на второй половине теста
    loss2, acc2 = model.evaluate(x_test2, y_test_cat[5000:], verbose=0)
    acc2_list.append(acc2)

# 6. Средняя точность по 100 эпохам
mean_acc1 = np.mean(acc1_list)
mean_acc2 = np.mean(acc2_list)

print(f"Средняя точность на первых 5000 примерах: {mean_acc1:.4f}")
print(f"Средняя точность на вторых 5000 примерах: {mean_acc2:.4f}")
print(f"Разница средней точности: {abs(mean_acc1 - mean_acc2):.4f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tqdm.notebook import tqdm

# 1. Загрузка и предобработка данных
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(-1, 28*28).astype('float32') / 255.0
X_test = X_test.reshape(-1, 28*28).astype('float32') / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Разделяем тест на две половины
X_test1, y_test1 = X_test[:5000], y_test[:5000]
X_test2, y_test2 = X_test[5000:], y_test[5000:]

# 3. Создаем модель
model = Sequential([
    layers.InputLayer(input_shape=(28*28,)),
    layers.Dense(500, activation='relu'),
    layers.Dense(500, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Компиляция с learning_rate=0.01
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-2)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# 5. Обучение 100 эпох по одной эпохе за раз и оценка на двух частях теста
val_accuracy1_history = []
val_accuracy2_history = []

for epoch_index in tqdm(range(100)):
    model.fit(X_train, y_train, epochs=1, verbose=0, batch_size=1024)
    val_loss1, val_accuracy1 = model.evaluate(X_test1, y_test1, verbose=0)
    val_loss2, val_accuracy2 = model.evaluate(X_test2, y_test2, verbose=0)

    val_accuracy1_history.append(val_accuracy1)
    val_accuracy2_history.append(val_accuracy2)

# 6. Средняя точность на двух тест-наборах и разница
mean_acc1 = np.mean(val_accuracy1_history)
mean_acc2 = np.mean(val_accuracy2_history)
print(f"Средняя точность на первых 5000 примерах: {mean_acc1:.4f}")
print(f"Средняя точность на вторых 5000 примерах: {mean_acc2:.4f}")
print(f"Разница средней точности (вторые - первые): {mean_acc2 - mean_acc1:.4f}")
```

```python
#Давай объединим все три задания (6.1, 6.2, 6.3) и визуализируем прогресс точностей:

#6.1 — отдельные модели → медиана точности

#6.2 — усреднение логитов → точность ансамбля

#6.3 — усреднение вероятностей → точность ансамбля

#Вот полный пример кода:
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# 1. Загрузка и предобработка данных
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0
y_train_cat = to_categorical(y_train, 10)

# 2. Параметры
num_models = 50
epochs_per_model = 10
batch_size = 128

# 3. Списки для результатов
accuracies_individual = []
all_predictions = []

# 4. Обучение 50 моделей (6.1)
for i in range(num_models):
    model = Sequential([
        layers.InputLayer(input_shape=(28*28,)),
        layers.Dense(500, activation='relu'),
        layers.Dense(500, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(x_train, y_train_cat, epochs=epochs_per_model, batch_size=batch_size, verbose=0)

    preds = model.predict(x_test, verbose=0)
    all_predictions.append(preds)

    test_acc = np.mean(np.argmax(preds, axis=1) == y_test)
    accuracies_individual.append(test_acc)

    tf.keras.backend.clear_session()
    del model, preds

accuracies_individual = np.array(accuracies_individual)
all_predictions = np.array(all_predictions)  # форма (50, 10000, 10)

# 5. Задание 6.2 — усреднение логитов
mean_logits = np.mean(all_predictions, axis=0)  # (10000,10)
pred_classes_logits = np.argmax(mean_logits, axis=1)
acc_ensemble_logits = np.mean(pred_classes_logits == y_test)

# 6. Задание 6.3 — усреднение вероятностей через softmax
mean_probs = np.array([tf.nn.softmax(x).numpy() for x in all_predictions]).mean(axis=0)
pred_classes_probs = np.argmax(mean_probs, axis=1)
acc_ensemble_probs = np.mean(pred_classes_probs == y_test)

# 7. Вывод результатов
print(f"Медианная точность отдельных моделей (6.1): {np.median(accuracies_individual):.4f}")
print(f"Точность ансамбля (усреднение логитов, 6.2): {acc_ensemble_logits:.4f}")
print(f"Точность ансамбля (усреднение вероятностей, 6.3): {acc_ensemble_probs:.4f}")

# 8. Визуализация
plt.figure(figsize=(8,5))
plt.bar(['Median Individual', 'Ensemble Logits', 'Ensemble Probabilities'],
        [np.median(accuracies_individual), acc_ensemble_logits, acc_ensemble_probs],
        color=['skyblue','orange','green'])
plt.ylabel('Accuracy')
plt.title('Сравнение точностей для 6.1, 6.2 и 6.3')
plt.ylim(0.95,1.0)
plt.show()
```

Объяснение:

accuracies_individual → точность каждой модели (6.1)

mean_logits → усреднение логитов всех моделей (6.2)

mean_probs → усреднение вероятностей через softmax (6.3)

Визуализация позволяет наглядно видеть рост точности с ансамблями моделей.

💡 Обычно:

Медиана отдельных моделей < усреднение логитов ≈ усреднение вероятностей

Ensemble повышает стабильность и точность.

```python
# -------------------------
# Импорты
# -------------------------
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import losses, optimizers
from tensorflow.keras.metrics import Precision, Recall

# -------------------------
# Пример данных
# -------------------------
num_features = 1024   # число признаков
num_classes = 10      # число классов
num_samples = 1000    # число обучающих примеров

# Случайные данные
X_train = np.random.rand(num_samples, num_features).astype(np.float32)
y_train = np.random.randint(0, num_classes, size=(num_samples,))
X_test = np.random.rand(200, num_features).astype(np.float32)
y_test = np.random.randint(0, num_classes, size=(200,))

# -------------------------
# Создание модели
# -------------------------
model = Sequential([
    Input(shape=(num_features,)),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(num_classes, activation='softmax')  # выходной слой соответствует числу классов
])

# -------------------------
# Компиляция модели
# -------------------------
model.compile(
    loss=losses.SparseCategoricalCrossentropy(from_logits=False),  # Softmax уже на выходе
    optimizer=optimizers.Adam(),
    metrics=['accuracy', Precision(), Recall()]
)

# -------------------------
# Обучение модели
# -------------------------
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=32
)

# -------------------------
# Проверка структуры модели
# -------------------------
model.summary()
```

### сделать ещё улучшенную версию, которая автоматически определяет, какой loss использовать (SparseCategoricalCrossentropy или CategoricalCrossentropy) в зависимости от формы меток.
универсальный вариант, который сам проверяет метки и подбирает правильный loss:

Если метки целые числа → SparseCategoricalCrossentropy

Если метки one-hot → CategoricalCrossentropy

✅ Преимущества этой версии:

Автоматически подбирает правильный loss в зависимости от формата меток.

Работает с целыми числами или one-hot.

Метрики accuracy, Precision, Recall корректно считаются.

Число выходных нейронов автоматически соответствует числу классов.

Минимизирует риск ошибки [Incompatible shapes].

```python
# -------------------------
# Импорты
# -------------------------
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import losses, optimizers
from tensorflow.keras.metrics import Precision, Recall

# -------------------------
# Предположим, у тебя уже есть реальные данные:
# X_train, y_train, X_test, y_test
# -------------------------

# -------------------------
# 1. Определяем формат меток и loss
# -------------------------
if y_train.ndim > 1 and y_train.shape[1] > 1:
    # Метки one-hot
    loss_fn = losses.CategoricalCrossentropy(from_logits=False)
    # В Keras метрики работают с числами, но можно оставить как есть для sparse
    y_train_labels = y_train
    y_test_labels = y_test
    print("Detected one-hot labels, using CategoricalCrossentropy")
else:
    # Метки в виде целых чисел
    loss_fn = losses.SparseCategoricalCrossentropy(from_logits=False)
    y_train_labels = y_train
    y_test_labels = y_test
    print("Detected integer labels, using SparseCategoricalCrossentropy")

# -------------------------
# 2. Определяем число признаков и классов
# -------------------------
num_features = X_train.shape[1]
num_classes = len(np.unique(y_train))  # для Sparse или Categorical

print("Number of features:", num_features)
print("Number of classes:", num_classes)
print("y_train shape:", y_train_labels.shape)
print("y_test shape:", y_test_labels.shape)

# -------------------------
# 3. Создание модели
# -------------------------
model = Sequential([
    Input(shape=(num_features,)),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(num_classes, activation='softmax')  # выходной слой = числу классов
])

# -------------------------
# 4. Компиляция модели
# -------------------------
model.compile(
    loss=loss_fn,
    optimizer=optimizers.Adam(),
    metrics=['accuracy', Precision(), Recall()]
)

# -------------------------
# 5. Обучение модели
# -------------------------
history = model.fit(
    X_train, y_train_labels,
    validation_data=(X_test, y_test_labels),
    epochs=10,
    batch_size=32
)

# -------------------------
# 6. Проверка структуры модели
# -------------------------
model.summary()
```
